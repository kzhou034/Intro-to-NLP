### 3.3: Implement the Naïve Bayes classifier to predict the emotion of a given text

```python
s = "As she hugged her daughter goodbye on the first day of college, she felt both sad to see her go and joyful knowing that she was embarking on a new and exciting chapter in her life."

s_tokens = word_tokenize(s.lower())
#print(s_tokens)

# from slides:
# scoring the test set:
# prior from training * ((# of times each word appears in the emotion set + 1) multiply them together)/(total words in emotion set with repeats + total count in vocab)

SadnessPrediction = 0
JoyPrediction = 0
FearPrediction = 0
AngerPrediction = 0
SurprisePrediction = 0
DisgustPrediction = 0

SadNumerator = 1
JoyNumerator = 1
FearNumerator = 1
AngerNumerator = 1
SurpriseNumerator = 1
DisgustNumerator = 1

SadDenominator = math.pow((len(AllSad) + len(vocab_set)), len(s_tokens))
JoyDenominator = math.pow((len(AllJoy) + len(vocab_set)), len(s_tokens))
FearDenominator = math.pow((len(AllFear) + len(vocab_set)), len(s_tokens))
AngerDenominator = math.pow((len(AllAnger) + len(vocab_set)), len(s_tokens))
SurpriseDenominator = math.pow((len(AllSurprise) + len(vocab_set)), len(s_tokens))
DisgustDenominator = math.pow((len(AllDisgust) + len(vocab_set)), len(s_tokens))

for token in s_tokens: #for each token in s_tokens
  word_appearances = 1
  for word_and_freq in SadTokenFreq_Copy: #for each tuple in the copied list
    if token in word_and_freq: #if the token is in that tuple
      #print(token)
      word_appearances += 1
  #print(token, word_appearances)
  SadNumerator *= word_appearances

  word_appearances = 1
  for word_and_freq in JoyTokenFreq_Copy: #for each tuple in the copied list
    if token in word_and_freq: #if the token is in that tuple
      #print(token)
      word_appearances += 1
  #print(token, word_appearances)
  JoyNumerator *= word_appearances

  word_appearances = 1
  for word_and_freq in FearTokenFreq_Copy: #for each tuple in the copied list
    if token in word_and_freq: #if the token is in that tuple
      #print(token)
      word_appearances += 1
  #print(token, word_appearances)
  FearNumerator *= word_appearances

  word_appearances = 1
  for word_and_freq in AngerTokenFreq_Copy: #for each tuple in the copied list
    if token in word_and_freq: #if the token is in that tuple
      #print(token)
      word_appearances += 1
  #print(token, word_appearances)
  AngerNumerator *= word_appearances

  word_appearances = 1
  for word_and_freq in SurpriseTokenFreq_Copy: #for each tuple in the copied list
    if token in word_and_freq: #if the token is in that tuple
      #print(token)
      word_appearances += 1
  #print(token, word_appearances)
  SurpriseNumerator *= word_appearances

  word_appearances = 1
  for word_and_freq in DisgustTokenFreq_Copy: #for each tuple in the copied list
    if token in word_and_freq: #if the token is in that tuple
      #print(token)
      word_appearances += 1
  #print(token, word_appearances)
  DisgustNumerator *= word_appearances

greatest_word = ''
greatest_prediction = 0

SadnessPrediction = SadnessPrior * (SadNumerator/SadDenominator)
greatest_word = 'Sadness'
greatest_prediction = SadnessPrediction
JoyPrediction = JoyPrior * (JoyNumerator/JoyDenominator)
if(JoyPrediction > greatest_prediction) :
  greatest_word = 'Joy'
  greatest_prediction = JoyPrediction

FearPrediction = FearPrior * (FearNumerator/FearDenominator)
if(FearPrediction > greatest_prediction) :
  greatest_word = 'Fear'
  greatest_prediction = FearPrediction

AngerPrediction = AngerPrior * (AngerNumerator/AngerDenominator)
if(AngerPrediction > greatest_prediction) :
  greatest_word = 'Anger'
  greatest_prediction = AngerPrediction

SurprisePrediction = SurprisePrior * (SurpriseNumerator/SurpriseDenominator)
if(SurprisePrediction > greatest_prediction) :
  greatest_word = 'Surprise'
  greatest_prediction = SurprisePrediction

DisgustPrediction = DisgustPrior * (DisgustNumerator/DisgustDenominator)
if(DisgustPrediction > greatest_prediction) :
  greatest_word = 'Disgust'
  greatest_prediction = DisgustPrediction

CalculatedPrediction = (greatest_word, greatest_prediction)

# print("Printing prior, num, denom: ")
# print("sad: ", SadnessPrior, SadNumerator, SadDenominator)
# print("joy: ", JoyPrior, JoyNumerator, JoyDenominator)
# print("fear: ", FearPrior, FearNumerator, FearDenominator)
# print("anger: ", AngerPrior, AngerNumerator, AngerDenominator)
# print("surprise: ", SurprisePrior, SurpriseNumerator, SurpriseDenominator)
# print("disgust: ", DisgustPrior, DisgustNumerator, DisgustDenominator)

print("Predictions for each emotion:")
print("Sadness prediction: ", SadnessPrediction)
print("Joy prediction: ", JoyPrediction)
print("Fear prediction: ", FearPrediction)
print("Anger prediction: ", AngerPrediction)
print("Surprise prediction: ", SurprisePrediction)
print("Disgust prediction: ", DisgustPrediction)
print("Predicted emotion for given sentence: ", CalculatedPrediction[0], CalculatedPrediction[1])
```

```python
#after we get the frequencies, we can start predicting

#Sadness 0
#Joy 1
#Fear 2
#Anger 3
#Surprise 4
#Disgust 5

# #get values for actual array
y_actual = []
i = -1
for sent in AllTestSentences:
  if sent is np.nan:
    continue
  i += 1
  if i < NumTestSent_Sad:
    y_actual.append(0)
  elif i < NumTestSent_Sad + NumTestSent_Joy:
    y_actual.append(1)
  elif i < NumTestSent_Sad + NumTestSent_Joy + NumTestSent_Fear:
    y_actual.append(2)
  elif i < NumTestSent_Sad + NumTestSent_Joy + NumTestSent_Fear + NumTestSent_Anger:
    y_actual.append(3)
  elif i < NumTestSent_Sad + NumTestSent_Joy + NumTestSent_Fear + NumTestSent_Anger + NumTestSent_Surprise:
    y_actual.append(4)
  elif i < NumTestSent_Sad + NumTestSent_Joy + NumTestSent_Fear + NumTestSent_Anger + NumTestSent_Surprise + NumTestSent_Disgust:
    y_actual.append(5)

#get values for predicted array
y_predicted = []
#print(AllTestSentences)
for sent in AllTestSentences:
  if sent is np.nan:
    continue

  #get curr sentence and tokenize it
  test_sent_tokens = sent

  # from slides:
  # scoring the test set:
  # prior from training * ((# of times each word appears in the emotion set + 1) multiply them together)/(total words in emotion set with repeats + total count in vocab)

  SadnessPrediction = 0
  JoyPrediction = 0
  FearPrediction = 0
  AngerPrediction = 0
  SurprisePrediction = 0
  DisgustPrediction = 0

  SadNumerator = 1
  JoyNumerator = 1
  FearNumerator = 1
  AngerNumerator = 1
  SurpriseNumerator = 1
  DisgustNumerator = 1

  SadDenominator = math.pow((len(AllSad) + len(vocab_set)), len(test_sent_tokens))
  JoyDenominator = math.pow((len(AllJoy) + len(vocab_set)), len(test_sent_tokens))
  FearDenominator = math.pow((len(AllFear) + len(vocab_set)), len(test_sent_tokens))
  AngerDenominator = math.pow((len(AllAnger) + len(vocab_set)), len(test_sent_tokens))
  SurpriseDenominator = math.pow((len(AllSurprise) + len(vocab_set)), len(test_sent_tokens))
  DisgustDenominator = math.pow((len(AllDisgust) + len(vocab_set)), len(test_sent_tokens))

  for token in test_sent_tokens: #for each token in test_sent_tokens
    word_appearances = 1
    for word_and_freq in SadTokenFreq: #for each tuple in the copied list
      if token in word_and_freq: #if the token is in that tuple
        #print(token)
        word_appearances += 1
    #print(token, word_appearances)
    SadNumerator *= word_appearances

    word_appearances = 1
    for word_and_freq in JoyTokenFreq: #for each tuple in the copied list
      if token in word_and_freq: #if the token is in that tuple
        #print(token)
        word_appearances += 1
    #print(token, word_appearances)
    JoyNumerator *= word_appearances

    word_appearances = 1
    for word_and_freq in FearTokenFreq: #for each tuple in the copied list
      if token in word_and_freq: #if the token is in that tuple
        #print(token)
        word_appearances += 1
    #print(token, word_appearances)
    FearNumerator *= word_appearances

    word_appearances = 1
    for word_and_freq in AngerTokenFreq: #for each tuple in the copied list
      if token in word_and_freq: #if the token is in that tuple
        #print(token)
        word_appearances += 1
    #print(token, word_appearances)
    AngerNumerator *= word_appearances

    word_appearances = 1
    for word_and_freq in SurpriseTokenFreq: #for each tuple in the copied list
      if token in word_and_freq: #if the token is in that tuple
        #print(token)
        word_appearances += 1
    #print(token, word_appearances)
    SurpriseNumerator *= word_appearances

    word_appearances = 1
    for word_and_freq in DisgustTokenFreq: #for each tuple in the copied list
      if token in word_and_freq: #if the token is in that tuple
        #print(token)
        word_appearances += 1
    #print(token, word_appearances)
    DisgustNumerator *= word_appearances

  greatest_word = ''
  greatest_prediction = 0

  SadnessPrediction = TestPrior_Sad * (SadNumerator/SadDenominator)
  greatest_word = 'Sadness'
  greatest_prediction = SadnessPrediction
  JoyPrediction = TestPrior_Joy * (JoyNumerator/JoyDenominator)
  if(JoyPrediction > greatest_prediction) :
    greatest_word = 'Joy'
    greatest_prediction = JoyPrediction

  FearPrediction = TestPrior_Fear * (FearNumerator/FearDenominator)
  if(FearPrediction > greatest_prediction) :
    greatest_word = 'Fear'
    greatest_prediction = FearPrediction

  AngerPrediction = TestPrior_Anger * (AngerNumerator/AngerDenominator)
  if(AngerPrediction > greatest_prediction) :
    greatest_word = 'Anger'
    greatest_prediction = AngerPrediction

  SurprisePrediction = TestPrior_Surprise * (SurpriseNumerator/SurpriseDenominator)
  if(SurprisePrediction > greatest_prediction) :
    greatest_word = 'Surprise'
    greatest_prediction = SurprisePrediction

  DisgustPrediction = TestPrior_Disgust * (DisgustNumerator/DisgustDenominator)
  if(DisgustPrediction > greatest_prediction) :
    greatest_word = 'Disgust'
    greatest_prediction = DisgustPrediction

  CalculatedPrediction = (greatest_word, greatest_prediction)

  # print("Printing prior, num, denom: ")
  # print("sad: ", TestPrior_Sad, SadNumerator, SadDenominator)
  # print("joy: ", TestPrior_Joy, JoyNumerator, JoyDenominator)
  # print("fear: ", TestPrior_Fear, FearNumerator, FearDenominator)
  # print("anger: ", TestPrior_Anger, AngerNumerator, AngerDenominator)
  # print("surprise: ", TestPrior_Surprise, SurpriseNumerator, SurpriseDenominator)
  # print("disgust: ", TestPrior_Disgust, DisgustNumerator, DisgustDenominator)

  # print("Predictions for each emotion:")
  # print("Sadness prediction: ", SadnessPrediction)
  # print("Joy prediction: ", JoyPrediction)
  # print("Fear prediction: ", FearPrediction)
  # print("Anger prediction: ", AngerPrediction)
  # print("Surprise prediction: ", SurprisePrediction)
  # print("Disgust prediction: ", DisgustPrediction)
  print("Predicted emotion for given sentence: ", CalculatedPrediction[0], CalculatedPrediction[1])

  if CalculatedPrediction[0] == 'Sadness':
    y_predicted.append(0)
  if CalculatedPrediction[0] == 'Joy':
    y_predicted.append(1)
  if CalculatedPrediction[0] == 'Fear':
    y_predicted.append(2)
  if CalculatedPrediction[0] == 'Anger':
    y_predicted.append(3)
  if CalculatedPrediction[0] == 'Surprise':
    y_predicted.append(4)
  if CalculatedPrediction[0] == 'Disgust':
    y_predicted.append(5)
```
